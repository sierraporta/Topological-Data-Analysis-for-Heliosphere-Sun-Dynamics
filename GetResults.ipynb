```python
def getstats(Var):
    cycle,mean,median,std,mini,maxi,var,odd_even=[],[],[],[],[],[],[],[]
    for j,i,k in zip(cycles,alldatas_inp,cycles2):
        mean.append(i[Var].mean())
        median.append(i[Var].median())
        std.append(i[Var].std())
        mini.append(i[Var].min())
        maxi.append(i[Var].max())
        var.append(Var)
        cycle.append(j)
        if k%2==0:
            odd_even.append("Even")
        else:
            odd_even.append("odd")
    dat=pd.DataFrame({"Cycle":cycle,"Odd-Even":odd_even,
                      "Variable":var,"Mean":mean,
                      "Median":median,"Std":std,
                      "Min":mini,"Max":maxi})
    dat.set_index("Cycle",inplace=True)
    return dat
```

```python
from scipy.stats import entropy
from scipy.signal import welch
import antropy as ant
import nolds
import hfda

def compute_entropies_and_fractal_dimensions(ts, sampling_rate=1.0):
    """
    Computes several entropy measures and fractal dimensions for a given time series.
    
    Parameters:
    - ts: array-like, the time series.
    - sampling_rate: float, the sampling frequency (needed for spectral entropy).

    Returns:
    - A dictionary with all computed measures.
    """
    # Entropy Measures
    shannon_entropy = entropy(np.histogram(ts, bins='auto', density=True)[0])
    sample_entropy = ant.sample_entropy(ts)
    permutation_entropy = ant.perm_entropy(ts, normalize=True)
    spectral_entropy = ant.spectral_entropy(ts, sf=sampling_rate, method='welch', normalize=True)
    app_entropy = ant.app_entropy(ts)
    k_max=5
    # Fractal Dimensions
    higuchi_fd = ant.higuchi_fd(ts)
    #higuchi_fd = hfda.measure(ts,k_max)
    katz_fd = ant.katz_fd(ts)
    petrosian_fd = ant.petrosian_fd(ts)
    lepel_ziv = ant.lziv_complexity(np.array(ts))
    dfa = ant.detrended_fluctuation(ts)
    return {
        "Shannon Entropy": shannon_entropy,
        "Sample Entropy": sample_entropy,
        "Permutation Entropy": permutation_entropy,
        "Spectral Entropy": spectral_entropy,
        "Approximate Entropy": app_entropy,
        "Higuchi Fractal Dimension": higuchi_fd,
        "Katz Fractal Dimension": katz_fd,
        "Petrosian Fractal Dimension": petrosian_fd,
        "Lempel-Ziv complexity": lepel_ziv,
        "Hurts exponent":dfa
    }
```

```python
def getstats2(Var):
    cycle,dfa,shannon,sample,permutation,espectral,appentropy,higuchi,katz,petrosian,lziv,var=[],[],[],[],[],[],[],[],[],[],[],[]
    for j,i in zip(cycles,alldatas_inp):
        a=compute_entropies_and_fractal_dimensions(i[Var])
        shannon.append(a["Shannon Entropy"])
        sample.append(a["Sample Entropy"])
        permutation.append(a["Permutation Entropy"])
        espectral.append(a["Spectral Entropy"])
        appentropy.append(a["Approximate Entropy"])
        higuchi.append(a["Higuchi Fractal Dimension"])
        katz.append(a["Katz Fractal Dimension"])
        petrosian.append(a["Petrosian Fractal Dimension"])
        lziv.append(a["Lempel-Ziv complexity"])
        dfa.append(a["Hurts exponent"])
        cycle.append(j)
    dat=pd.DataFrame({"Cycle":cycle,"Shannon Entropy":shannon,
                      "Sample Entropy":sample,"Permutation Entropy":permutation,
                      "Spectral Entropy":espectral,"Approximate Entropy":appentropy,
                      "Higuchi Fractal Dimension":higuchi,"Katz Fractal Dimension":katz,
                      "Petrosian Fractal Dimension":petrosian,"Lempel-Ziv complexity":lziv,
                     "Hurts exponent":dfa})
    dat.set_index("Cycle",inplace=True)
    return dat
```

### Pairwise correlations: 

Matrix plot of pairwise correlations (Pearson r) between all eleven complexity metrics. Do this within each physical variable and then on the pooled set. The goal is to verify that clusters we anticipatedâ€”e.g. the tight Approximate/Sampleâ€entropy pair, or the Katzâ€“LZ blockâ€”are present in the real data. A single 
11x11 heat-map per variable, stacked vertically, is enough; colour bars aligned lets the eye pick out which variables are intrinsically â€œmore chaoticâ€.

**Why?** This validates that the metrics convey complementaryâ€”not redundantâ€”information and sets up the need for dimension reduction in the next step.

```python
def get_pairwise_corr(var,name,save=False):
    # List of complexityâ€‘metric columns expected in the file
    metric_candidates = [
        'Shannon Entropy', 'Sample Entropy', 'Permutation Entropy', 'Spectral Entropy', 'Approximate Entropy', 
        'Higuchi Fractal Dimension', 'Katz Fractal Dimension', 'Petrosian Fractal Dimension',
        'Lempel-Ziv complexity', 'Hurts exponent']
    # Var is one of:
    # ['Scalar B, nT', 'SW Plasma Temperature, K', 'SW Proton Density, N/cm^3', 'SW Plasma Speed, km/s',
    #   'Alpha/Prot. ratio', 'Flow pressure', 'Kp index', 'R (Sunspot)', 'Dst-index', 'f10.7_index']
    if var=="None":
        df=data_allvar.copy()
    else:
        df=data_allvar[data_allvar["Variable"]==var].copy()
    
    # Keep only the columns that exist in the dataframe
    metric_cols = [col for col in metric_candidates if col in df.columns]
    # Calculate the correlation matrix (Pearson r)
    corr_matrix = df[metric_cols].corr(method='pearson')
    # Plot the correlation heatâ€‘map using matplotlib
    fig, ax = plt.subplots(figsize=(9, 5))
    cax = ax.imshow(corr_matrix, vmin=-1, vmax=1, cmap='Spectral', aspect=0.6)
    # Ticks & labels
    ax.set_xticks(np.arange(len(metric_cols)))
    ax.set_yticks(np.arange(len(metric_cols)))
    ax.set_xticklabels(['Shannon \n Entropy', 'Sample \n Entropy', 'Permutation \n Entropy', 'Spectral \n Entropy', 'Approximate \n Entropy', 
        'Higuchi \n Fractal \n Dimension', 'Katz \n Fractal \n Dimension', 'Petrosian \n Fractal \n Dimension',
        'Lempel-Ziv \n complexity', 'Hurts \n exponent'], rotation=90, ha='center',fontsize=10)
    ax.set_yticklabels(metric_cols,fontsize=10)
    # Add correlation values on top of cells
    for i in range(len(metric_cols)):
        for j in range(len(metric_cols)):
            ax.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}",
                    ha='center', va='center', color='black', fontsize=10)
    # Add colorbar
    cbar=fig.colorbar(cax, ax=ax, label='Pearson correlation (r)',shrink=0.8)
    # Change the font size of the colorbar tick labels
    cbar.ax.tick_params(labelsize=10, size=10)
    cbar.set_label(r'Pearson correlation ($r$)', size=10)
    if var=="None":
        ax.set_title(r"Pairwise Pearson correlations between complexity metrics for all variables", fontsize=10)
    else:
        ax.set_title(r"Pairwise Pearson correlations between complexity metrics for {}".format(var), fontsize=10)
    plt.tight_layout()
    if save:
        plt.savefig("figures/Correlations_{}".format(name),dpi=300,bbox_inches='tight')
    return plt.show()
```

### Principal Component Analysis...

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from matplotlib.lines import Line2D
# === Load data (adjust path if needed) ===
df = data_allvar.copy()
df["CyclePhase"]=df.index
df=df.reset_index()

# === Identify metric columns ===
metric_cols = [
        'Shannon Entropy', 'Sample Entropy', 'Permutation Entropy', 'Spectral Entropy', 'Approximate Entropy', 
        'Higuchi Fractal Dimension', 'Katz Fractal Dimension', 'Petrosian Fractal Dimension',
        'Lempel-Ziv complexity', 'Hurts exponent']

metric_cols = [c for c in metric_cols if c in df.columns]  # keep existing

# --- Numerical pipeline: impute â†’ scale â†’ PCA --------------------------
numeric_pipeline = Pipeline(steps=[
        ("imputer",  SimpleImputer(strategy='mean')),
        ("scaler",   StandardScaler()),
        ("pca",      PCA(n_components=3, random_state=42))
])

X_pca = numeric_pipeline.fit_transform(df[metric_cols])

var_ratio = numeric_pipeline.named_steps["pca"].explained_variance_ratio_
print(f"PC1 explains {var_ratio[0]*100:.1f} %, PC2 {var_ratio[1]*100:.1f} %, PC3 {var_ratio[2]*100:.1f} %")

# ---  obtener los loadings  ----------------------------------------------
pca_model = numeric_pipeline.named_steps["pca"]          # alias corto
loadings  = pd.DataFrame(
    pca_model.components_.T,                             # mÃ©tricas en filas
    index=metric_cols,
    columns=[f"PC{i+1}" for i in range(pca_model.n_components_)]
)

print("\nLoadings (coeficientes de cada mÃ©trica en las 3 PC):")
print(loadings.round(3))

# (opcional) guardar a CSV para incluirla en LaTeX
#loadings.to_csv("PCA_loadings.csv")

# --- 3. Prepare labels for colour / marker -------------------------------
if 'CyclePhase' in df.columns:
    cycles = df['CyclePhase'].str.extract(r'(\d+)', expand=False)
    phases = df['CyclePhase'].str.extract(r'([ud])', expand=False)
else:
    cycles = df.get('Cycle', pd.Series(np.zeros(len(df)), index=df.index))
    phases = df.get('Phase', pd.Series(['u']*len(df), index=df.index))

unique_cycles = sorted(cycles.unique(), key=int)
cycle_colors = dict(zip(unique_cycles, plt.cm.tab10.colors[:len(unique_cycles)]))
marker_dict = {'u': 'o', 'd': '^'}

# --- 4. Scatter plot ------------------------------------------------------
fig, ax = plt.subplots(figsize=(7, 5))
for i in range(len(df)):
    col = cycle_colors[cycles.iloc[i]]
    mk  = marker_dict.get(phases.iloc[i], 'o')
    ax.scatter(X_pca[i, 0], X_pca[i, 1],
               color=col, marker=mk, s=60, edgecolor='k', alpha=0.8)

ax.set_xlabel(f"PC1 ({var_ratio[0]*100:.1f} \%)", fontsize=10)
ax.set_ylabel(f"PC2 ({var_ratio[1]*100:.1f} \%)", fontsize=10)
ax.set_title("PCA of complexity metrics", fontsize=10)

# legend handles
handles = [Line2D([0], [0], color=cycle_colors[c], marker='o', markersize=8, linestyle='',
                  label=f"Cycle {c}") for c in unique_cycles]
handles += [Line2D([0], [0], color='grey', marker=m, markersize=8, linestyle='',
                   label=f"Phase {p}") for p, m in marker_dict.items()]
ax.legend(handles=handles, bbox_to_anchor=(1.02, 1), loc='upper left', frameon=False, fontsize=10)
ax.tick_params(labelsize=11)
plt.ylim(-2,2)
plt.xlim(-6,4)
plt.grid(alpha=0.5)
plt.tight_layout()
plt.savefig("figures/PCAloadings",dpi=300,bbox_inches='tight')
plt.show()
```

```python
import numpy as np
from itertools import combinations

def cliffs_delta(x, y):
    """
    Cliff's delta for two paired or independent samples.
    Î´ = (n_pos - n_neg) / (n_x * n_y)
    """
    x = np.asarray(x)
    y = np.asarray(y)
    n_x = len(x)
    n_y = len(y)
    n_pos = np.sum(x[:, None] > y)   # x_i > y_j
    n_neg = np.sum(x[:, None] < y)   # x_i < y_j
    delta = (n_pos - n_neg) / (n_x * n_y)
    return delta

def boot_ci(data, func=np.mean, n_boot=10000, seed=42, alpha=0.05):
    rng = np.random.default_rng(seed)
    boots = [func(rng.choice(data, size=len(data), replace=True))
             for _ in range(n_boot)]
    lo = np.percentile(boots, 100*alpha/2)
    hi = np.percentile(boots, 100*(1-alpha/2))
    return lo, hi
```

```python
# Wilcoxon + bootstrap BCa CI + Cliffâ€™s Î´
# --------------------------------------

import numpy as np
import pandas as pd
from scipy.stats import wilcoxon
from pingouin import compute_bootci
from statsmodels.stats.multitest import multipletests

# Identifying columns ------------------------------------------------------
cycle_col = "Cycle"       # integer 20â€“25
phase_col = "Phase"       # "u" / "d"
var_col   = "Variable"    # physical variable

# Complexity metrics already defined
metric_cols = [c for c in metric_cols if c in df.columns]

records = []

for variable, sub in df.groupby(var_col):
    for metric in metric_cols:
        # tablas u-d por ciclo
        pivot  = sub.pivot_table(index=cycle_col,
                                 columns=phase_col,
                                 values=metric,
                                 aggfunc="mean")
        paired = pivot.dropna()          # ciclos con ambas fases
        if paired.shape[0] < 2:
            continue                    # Wilcoxon necesita â‰¥2 pares

        # diferencias u âˆ’ d
        diff = paired["u"] - paired["d"]
        delta = diff.mean()

        # BCa 95â€‰% bootstrap CI (10â€‰000 remuestreos estratificados)
        ci_low, ci_high = boot_ci(diff.values, func=np.mean, n_boot=10000)

        # Cliffâ€™s Î´
        cd = cliffs_delta(paired['u'].values, paired['d'].values)

        # Wilcoxon exacto
        stat, p = wilcoxon(paired["u"], paired["d"],
                           zero_method="wilcox")

        # almacenar
        records.append(dict(Variable=variable,
                            Metric=metric,
                            n_pairs=len(paired),
                            val_u=paired["u"].mean(),
                            val_d=paired["d"].mean(),
                            Delta=delta,
                            CI_low=ci_low,
                            CI_high=ci_high,
                            Cliff_delta=cd,
                            W=stat,
                            pval=p))

# Convertir a DataFrame
results = pd.DataFrame(records)

# CorrecciÃ³n FDR (Benjaminiâ€“Hochberg)
results["qval"] = multipletests(results["pval"],
                                method="fdr_bh")[1]

# Mostrar un resumen ordenado por q
cols = ["Variable", "Metric", "n_pairs", "val_u", "val_d",
        "Delta", "CI_low", "CI_high", "Cliff_delta",
        "W", "pval", "qval"]
results.sort_values("qval")[cols].head(15)
```

### Oddâ€“even solar-cycle asymmetry
Construct a linear mixed-effects model for each metric:

$$Metric_{ijk}=ð›½0+ð›½1(OddEven_{i})+ð›½2(Phase_{j})+ð‘¢_{ð‘˜}+ðœ€_{ijk}$$

```python
import numpy as np, random, pandas as pd
import statsmodels.formula.api as smf
from statsmodels.stats.multitest import multipletests

# ---------- funciones -----------------------------------------------------
def perm_pvalue(data, formula, B=10000, seed=42):
    """Permutation p-value for Î²_OddEven, keeping phase & cycle structure."""
    random.seed(seed)
    np.random.seed(seed)
    # Î² observado
    beta_obs = smf.ols(formula, data=data).fit(cov_type='HC3').params["OddEvenNum"]
    count = 0
    for _ in range(B):
        shuffled = data.copy()
        # baraja paridad dentro de cada ciclo-fase
        shuffled["OddEvenNum"] = (
            shuffled.groupby(["Cycle", "Phase"]).OddEvenNum.transform(
                np.random.permutation)
        )
        beta_perm = smf.ols(formula, data=shuffled)\
                       .fit(cov_type='HC3').params["OddEvenNum"]
        if abs(beta_perm) >= abs(beta_obs):
            count += 1
    return (count + 1) / (B + 1)     # p_perm

# ---------- preparaciÃ³n de datos -----------------------------------------
df["Cycle"]      = pd.to_numeric(df["Cycle"], errors="coerce").astype(int)
df["OddEvenNum"] = (df["Cycle"] % 2).astype(int)
df["PhaseNum"]   = df["Phase"].map({"u": 0, "d": 1})

metric_cols = [
    "Shannon Entropy", "Sample Entropy", "Approximate Entropy",
    "Permutation Entropy", "Spectral Entropy",
    "Higuchi Fractal Dimension", "Katz Fractal Dimension",
    "Petrosian Fractal Dimension", "Lempel-Ziv complexity",
    "Hurts exponent"
]
metric_cols = [m for m in metric_cols if m in df.columns]

rows = []
for metric in metric_cols:
    tmp = df[[metric, "OddEvenNum", "PhaseNum", "Variable",
              "Cycle", "Phase"]].dropna()
    if len(tmp) < 20:
        continue
    formula = f'Q("{metric}") ~ OddEvenNum + PhaseNum + C(Variable)'
    model   = smf.ols(formula, data=tmp).fit(cov_type='HC3')
    p_perm  = perm_pvalue(tmp, formula, B=10000, seed=42)
    rows.append({
        "Metric":    metric,
        "Î²_OddEven": model.params["OddEvenNum"],
        "p_perm":    p_perm,
        "Î²_Phase":   model.params["PhaseNum"],
        "p_Phase":   model.pvalues["PhaseNum"]
    })

parity_df = pd.DataFrame(rows)
parity_df["q_perm"] = multipletests(parity_df["p_perm"], method='fdr_bh')[1]

parity_df.sort_values("q_perm")
```
