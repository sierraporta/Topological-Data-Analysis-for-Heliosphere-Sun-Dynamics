```python
from sklearn.preprocessing import StandardScaler, QuantileTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
# ─────────────────────────────────────────────
# 1. Pre-procesamiento
# ─────────────────────────────────────────────
X = data.drop(columns=[var]).values
y = data[var].values.reshape(-1, 1)

#sc_x, sc_y = StandardScaler(), StandardScaler()
#X_std = sc_x.fit_transform(X).astype(np.float32)
#y_std = sc_y.fit_transform(y).astype(np.float32)
sc_x, sc_y = QuantileTransformer(random_state=rs), QuantileTransformer(random_state=rs)
X_std = sc_x.fit_transform(X).astype(np.float32)
y_std = sc_y.fit_transform(y).astype(np.float32)

X_tr, X_te, y_tr, y_te = train_test_split(X_std, 
                                          y_std, 
                                          test_size=tsize, 
                                          random_state=rs)
```

# 1 | Multiple Layer Perceptron - GPU

```python
# ================================================================
#  MLP con PyTorch + Optuna para búsqueda de hiper-parámetros
#  Dataset: 4 602 × 14   ‒ Target: R
#  CUDA se usa si está disponible (MX250)
#  Requiere: pip install optuna torch torchvision scikit-learn
# ================================================================
import numpy as np, torch, torch.nn as nn, optuna, joblib
from torch.utils.data import TensorDataset, DataLoader

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ─────────────────────────────────────────────
# 2. Definición del modelo
# ─────────────────────────────────────────────
class MLP(nn.Module):
    def __init__(self, in_dim: int,
                 h1: int, h2: int | None, h3: int | None,
                 p_drop: float = 0.2):
        """Hasta 3 capas ocultas; las que tengan None se omiten."""
        super().__init__()
        layers = [nn.Linear(in_dim, h1), nn.ReLU(), nn.Dropout(p_drop)]
        if h2:
            layers += [nn.Linear(h1, h2), nn.ReLU(), nn.Dropout(p_drop)]
            prev = h2
        else:
            prev = h1
        if h3:
            layers += [nn.Linear(prev, h3), nn.ReLU(), nn.Dropout(p_drop)]
            prev = h3
        layers += [nn.Linear(prev, 1)]
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

# ─────────────────────────────────────────────
# 3. Función de entrenamiento con early-stopping
# ─────────────────────────────────────────────
def train_mlp(params, epochs=500, patience=20):
    model = MLP(
        in_dim=X_tr.shape[1],
        h1=params['h1'],
        h2=params.get('h2'),
        h3=params.get('h3'),
        p_drop=params['dropout']).to(device)

    opt = torch.optim.Adam(
        model.parameters(),
        lr=params['lr'],
        weight_decay=params['weight_decay'])

    batch_size = params['batch']
    ds_train   = TensorDataset(torch.from_numpy(X_tr), torch.from_numpy(y_tr))
    dl_train   = DataLoader(ds_train, batch_size=batch_size, shuffle=True)

    best_val, wait = np.inf, 0
    for _ in range(epochs):
        # ── forward/backward ───────────────────────────────────
        model.train()
        for xb, yb in dl_train:
            xb, yb = xb.to(device), yb.to(device)
            opt.zero_grad()
            loss = nn.functional.mse_loss(model(xb), yb)
            loss.backward()
            opt.step()

        # ── validación rápida ─────────────────────────────────
        model.eval()
        with torch.no_grad():
            preds = model(torch.from_numpy(X_te).to(device)).cpu().numpy()
            val_loss = mean_squared_error(y_te, preds)
        # early stop
        if val_loss < best_val - 1e-4:
            best_val = val_loss
            best_state = model.state_dict()
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                break

    model.load_state_dict(best_state)
    return model, best_val

# ─────────────────────────────────────────────
# 4. Objetivo Optuna
# ─────────────────────────────────────────────
def objective(trial: optuna.Trial) -> float:
    # muestreo de hiper-parámetros
    h1 = trial.suggest_int('h1',       32, 128, step=32)
    h2 = trial.suggest_int('h2',       16, 128, step=16)
    use_h3 = trial.suggest_categorical('use_h3', [False, True])
    h3 = trial.suggest_int('h3', 16, 64, step=16) if use_h3 else None

    params = dict(
        h1       = h1,
        h2       = h2,
        h3       = h3,
        dropout  = trial.suggest_float('dropout', 0.0, 0.4, step=0.1),
        lr       = trial.suggest_float('lr', 1e-4, 5e-3, log=True),
        weight_decay = trial.suggest_float('weight_decay', 1e-5, 1e-2, log=True),
        batch    = trial.suggest_categorical('batch', [64, 128, 256]),
    )

    # entrena y evalúa
    _, val_rmse = train_mlp(params)
    # Optuna minimiza -> devolver RMSE
    return val_rmse

# ─────────────────────────────────────────────
# 5. Ejecución del estudio
# ─────────────────────────────────────────────
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=40, timeout=None, show_progress_bar=True)

print("\n⇢  Mejor RMSE válido  :", study.best_value**0.5)   # pasa de MSE a RMSE
print("⇢  Mejores parámetros :", study.best_params)

joblib.dump(study, "optuna_mlp_study.pkl")

# ─────────────────────────────────────────────
# 6. Entrenamiento final con el mejor conjunto
# ─────────────────────────────────────────────
best_model, _ = train_mlp(study.best_params, epochs=800, patience=30)
```

```python
# ── Métricas finales en escala original ──────
with torch.no_grad():
    best_model.eval()
    preds_te_std = best_model(torch.from_numpy(X_te).to(device)).cpu().numpy()
    preds_tr_std = best_model(torch.from_numpy(X_tr).to(device)).cpu().numpy()

y_te_inv  = sc_y.inverse_transform(y_te)
y_tr_inv  = sc_y.inverse_transform(y_tr)
preds_te  = sc_y.inverse_transform(preds_te_std)
preds_tr  = sc_y.inverse_transform(preds_tr_std)

print("R² test :", r2_score(y_te_inv, preds_te))
print("RMSE test:", mean_squared_error(y_te_inv, preds_te, squared=False))
print("R² train:", r2_score(y_tr_inv, preds_tr))
print("RMSE train:", mean_squared_error(y_tr_inv, preds_tr, squared=False))
```

# 3 | XGBoost GPU

```python
import xgboost as xgb
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
import optuna, numpy as np

dtrain, dtest = xgb.DMatrix(X_tr, label=y_tr), xgb.DMatrix(X_te, label=y_te)

# ── ¿XGBoost tiene GPU? ───────────────────────────────
def xgb_gpu_available() -> bool:
    try:
        # Desde XGB 2.0 existe esto:
        return xgb.core._has_cuda_support()
    except AttributeError:
        # Fallback: intenta una llamada dummy
        try:
            xgb.train({'tree_method':'gpu_hist'}, dtrain, num_boost_round=1, verbose_eval=False)
            return True
        except xgb.core.XGBoostError:
            return False

USE_GPU = xgb_gpu_available()
print("XGBoost GPU disponible?:", USE_GPU)

tree_method = 'gpu_hist' if USE_GPU else 'hist'

# ── Optuna objective ──────────────────────────────────

def objective(trial):
    param = {
        "objective": "reg:squarederror",
        "tree_method": tree_method,          # 'gpu_hist' o 'hist'
        "learning_rate": trial.suggest_float("eta", 0.02, 0.15, log=True),
        "max_depth": trial.suggest_int("max_depth", 1, 4),
        "subsample": trial.suggest_float("subsample", 0.5, 0.9),
        "colsample_bytree": trial.suggest_float("colsample", 0.4, 0.9),
        "lambda": trial.suggest_float("lambda", 1.0, 10.0, log=True),
        "alpha": trial.suggest_float("alpha", 0.0, 1.0),
        "min_child_weight": trial.suggest_float("min_child_weight", 5, 10, log=True),
    }
    n_rounds = trial.suggest_int("n_rounds", 500, 3000, step=500)

    booster = xgb.train(
        param,
        dtrain,
        num_boost_round=n_rounds,
        evals=[(dtest, "val")],
        verbose_eval=False,
        early_stopping_rounds=50,
    )
    preds = booster.predict(dtest)
    return mean_squared_error(y_te, preds)          # Optuna minimiza

study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=50, show_progress_bar=True, timeout=None)

best_params = study.best_params | {
    "objective": "reg:squarederror",
    "tree_method": tree_method
}
best_rounds = best_params.pop("n_rounds")
booster = xgb.train(best_params, dtrain,
                    num_boost_round=best_rounds,
                    evals=[(dtest, "val")],
                    early_stopping_rounds=50,
                    verbose_eval=False)
```

```python
preds_te_std2 = booster.predict(dtest)
preds_tr_std2 = booster.predict(dtrain)

y_te_inv2  = sc_y.inverse_transform(y_te)
y_tr_inv2  = sc_y.inverse_transform(y_tr)
preds_te2  = sc_y.inverse_transform(np.array(preds_te_std2).reshape(-1, 1))
preds_tr2  = sc_y.inverse_transform(np.array(preds_tr_std2).reshape(-1, 1))

print("R² test :", r2_score(y_te_inv2, preds_te2))
print("RMSE test:", mean_squared_error(y_te_inv2, preds_te2, squared=False))
print("R² train:", r2_score(y_tr_inv2, preds_tr2))
print("RMSE train:", mean_squared_error(y_tr_inv2, preds_tr2, squared=False))
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.title("Predictions")
plt.scatter(y_tr_inv2,preds_tr2,label="Train")
plt.scatter(y_te_inv2,preds_te2,label="Test")
plt.legend()
plt.subplot(122)
plt.title("Resisuals")
plt.scatter(y_tr_inv2,preds_tr2-y_tr_inv2,label="Train")
plt.scatter(y_te_inv2,preds_te2-y_te_inv2,label="Test")
plt.legend()
plt.show()
```

```python
import numpy as np, matplotlib.pyplot as plt, xgboost as xgb, torch
from sklearn.inspection import permutation_importance
from sklearn.metrics import mean_squared_error
from typing import Any

# ═══════════════════════════════════════════════════════════════════
# 1. Wrappers: dan predict() y un fit() vacío (requisito sklearn)
# ═══════════════════════════════════════════════════════════════════
class TorchRegressorWrapper:
    def __init__(self, model: torch.nn.Module, scaler_y):
        self.model = model.eval().to(next(model.parameters()).device)
        self.scaler_y = scaler_y
    def fit(self, X: Any, y: Any):     # <- para que sklearn no proteste
        return self
    def predict(self, X):
        X_t = torch.tensor(X, dtype=torch.float32, device=next(self.model.parameters()).device)
        with torch.no_grad():
            y_hat_std = self.model(X_t).cpu().numpy()
        return self.scaler_y.inverse_transform(y_hat_std).ravel()

class XGBRegressorWrapper:
    def __init__(self, booster: xgb.Booster):
        self.bst = booster
    def fit(self, X: Any, y: Any):
        return self
    def predict(self, X):
        return self.bst.predict(xgb.DMatrix(X))

mlp_wrap = TorchRegressorWrapper(best_model, sc_y)
xgb_wrap = XGBRegressorWrapper(booster)

# ═══════════════════════════════════════════════════════════════════
# 2. Conjuntos para permutar
#    · MLP → entradas estandarizadas
#    · XGB → sin escalar
# ═══════════════════════════════════════════════════════════════════
X_te_std = sc_x.transform(X_te)        # estándar usado en entrenamiento MLP

# ═══════════════════════════════════════════════════════════════════
# 3. Permutation Feature Importance
# ═══════════════════════════════════════════════════════════════════
def pfi(estimator, X, y):
    return permutation_importance(
        estimator, X, y,
        n_repeats=50,
        scoring="neg_root_mean_squared_error",
        random_state=rs,
        n_jobs=-1)

pi_mlp = pfi(mlp_wrap, X_te, y_te)
pi_xgb = pfi(xgb_wrap, X_te, y_te)
```
